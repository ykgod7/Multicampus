```python
num = 121

def get_largest(a):
    i = 1
    numbers = list()
    
    while i < a:
        if a % i == 0:
            numbers.append(i)
        i+=1

    i = 1
    max_length = len(numbers)
    
    while i < max_length:
        current = numbers[i] 
        j = 1
        i += 1 # i = 4
        
        while j < len(numbers):
            next_num = numbers[j] 
            
            if current == next_num:
                j+=1
            elif current % next_num == 0:
                numbers.remove(current) 
                max_length = len(numbers)
                i -= 1
                break
            else:
                j+=1
                    
    print(numbers[max_length-1])
       
        
get_largest(num)
```

```python
def match_number(num_1, num_2):
    pair_number = list()
    
    while num_1 < 999:
        num_2 = 100
        while num_2 < 999:
            multi_num = num_1 * num_2
            num_length = len(str(multi_num))
            
            if num_length % 2 == 0:
                pair_1 = str(multi_num)[0:3]
                pair_2 = str(multi_num)[-1:-4:-1]
                if pair_1 == pair_2:
                    pair_number.append(multi_num)
                    num_2 += 1
                else:
                    num_2 += 1
            else:
                num_2 += 1 
                
            
        num_1 += 1
    print(pair_number[-1])
    
match_number(100,100)
```
